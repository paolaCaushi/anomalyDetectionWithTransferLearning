```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, precision_recall_fscore_support, classification_report
import matplotlib.pyplot as plt

# Define the Online Autoencoder class
class OnlineAutoencoder(nn.Module):
    def __init__(self, input_dim):
        super(OnlineAutoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32)
        )
        self.decoder = nn.Sequential(
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim),
            nn.Sigmoid()
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

# Load and preprocess the dataset
data_file_2 = "Book1.xlsx"  # Replace with the path to your dataset
df_2 = pd.read_excel(data_file_2, sheet_name="Sheet1", header=0)

# Map columns and create target variable
column_mapping = {
    'Codice': 'idAnomalia',
    'CodiceTipo': 'note',
    'DataOraRicezione': 'dataAnomalia',
    'Latitudine': 'latitudine',
    'Longitudine': 'longitudine',
    'DataOraRilevazione': 'dataCoordinate',
    'DataOraCreazione': 'dataSegnalazione',
    'Semirimorchio': 'idConcentratore'
}
df_2 = df_2.rename(columns=column_mapping)
df_2['Target'] = df_2['note'].apply(lambda x: 1 if x == 6 else 0)  # Adjust target condition if needed
datetime_columns_2 = ['dataAnomalia', 'dataCoordinate', 'dataSegnalazione']
for col in datetime_columns_2:
    df_2[col] = pd.to_datetime(df_2[col], errors='coerce').astype('int64') // 10**9
for col in datetime_columns_2:
    df_2[col].fillna(df_2[col].median(), inplace=True)

# Extract features and target
X_2 = df_2.drop(columns=['Target'])
y_2 = df_2['Target']

# Convert categorical features to one-hot encoding
X_2 = pd.get_dummies(X_2)

# Ensure all columns are numeric before scaling
X_2 = X_2.apply(pd.to_numeric, errors='coerce')  # Convert all columns to numeric, invalid parsing will be set as NaN
X_2.fillna(0, inplace=True)  # Fill NaN values with zero or another strategy

# Normalize the features
scaler = StandardScaler()
X_2 = scaler.fit_transform(X_2)

# Convert to torch tensors
X_2_tensor = torch.tensor(X_2, dtype=torch.float32)
y_2_tensor = torch.tensor(y_2.values, dtype=torch.float32)

# Prepare data for training
train_dataset = TensorDataset(X_2_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Initialize and train the Online Autoencoder
input_dim_2 = X_2.shape[1]
online_autoencoder = OnlineAutoencoder(input_dim_2)
criterion = nn.MSELoss()
optimizer = optim.Adam(online_autoencoder.parameters(), lr=0.001)
num_epochs = 50

# Training loop
online_autoencoder.train()
for epoch in range(num_epochs):
    for data in train_loader:
        inputs = data[0]
        optimizer.zero_grad()
        outputs = online_autoencoder(inputs)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()
    print(f'Online Autoencoder Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Evaluate the Online Autoencoder
online_autoencoder.eval()
with torch.no_grad():
    X_2_tensor = torch.tensor(X_2, dtype=torch.float32)
    reconstructed_outputs = online_autoencoder(X_2_tensor)

    # Compute per-sample reconstruction error
    reconstruction_errors = torch.mean((reconstructed_outputs - X_2_tensor) ** 2, dim=1)
    reconstruction_error = reconstruction_errors.numpy()

# Use reconstruction error as an anomaly score
threshold = np.percentile(reconstruction_error, 95)  # 95th percentile threshold
anomaly_predictions = (reconstruction_error > threshold).astype(int)

# Evaluate the performance
accuracy_online = accuracy_score(y_2, anomaly_predictions)
roc_auc_online = roc_auc_score(y_2, reconstruction_error)
precision_online, recall_online, f1_online, _ = precision_recall_fscore_support(y_2, anomaly_predictions, average='binary')

print(f'Online Autoencoder Accuracy: {accuracy_online}')
print(f'Online Autoencoder ROC-AUC Score: {roc_auc_online}')
print(f'Online Autoencoder Precision: {precision_online}')
print(f'Online Autoencoder Recall: {recall_online}')
print(f'Online Autoencoder F1-Score: {f1_online}')
print("Online Autoencoder Classification Report:")
print(classification_report(y_2, anomaly_predictions))

# Add predictions to the dataset for further analysis
df_2['AnomalyPrediction'] = anomaly_predictions
df_2['AnomalyScore'] = reconstruction_error

# Plot the distribution of anomaly scores
plt.hist(df_2['AnomalyScore'], bins=50, alpha=0.75)
plt.title('Distribution of Anomaly Scores')
plt.xlabel('Anomaly Score')
plt.ylabel('Frequency')
plt.show()

# Print the analysis results
anomaly_count = df_2['AnomalyPrediction'].sum()
total_count = len(df_2)
anomaly_ratio = anomaly_count / total_count

print(f'Total records: {total_count}')
print(f'Predicted anomalies: {anomaly_count}')
print(f'Anomaly ratio: {anomaly_ratio:.4f}')
```
